```python

# 파일 경로 설정
filepath =  "/content/drive/MyDrive/edf.edf"

# EDF 파일 읽기
with pyedflib.EdfReader(filepath) as reader:
    # 정보 출력
    print("EDF 파일 정보:")
    # print("  파일 버전:", reader.file_version)
    # print("  데이터 타입:", reader.data_format)
    # print("  샘플링 주파수:", reader.getSampleFrequencies()[0])

    # 데이터 읽기
    signals = reader.readSignal(chn=0)  # 특정 채널 선택

    # 데이터 처리 및 분석
```

```python
signals1 = signals[:1000]
```

```python
import matplotlib.pyplot as plt
from scipy import signal

# ... (데이터 읽기 및 전처리 코드)

# 스펙트럼 분석
Pxx, freqs = signal.welch(signals1, nperseg=1024, noverlap=8)

# 시각화
plt.figure(figsize=(10, 6))
plt.loglog(freqs, Pxx)
plt.xlabel("Frequency")
plt.ylabel("Power Spectral Density")
plt.show()
```
